```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
train = pd.read_csv('Train_Data.csv')
test = pd.read_csv('Test_Data.csv')
```

```python
train.head()
```

```python
train['age_group'] = train['age_group'].map({'Adult': 0, 'Senior': 1})
train.head()
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='median')
train[train.columns] = imputer.fit_transform(train)

train.isnull().sum()
```

```python
x = train.drop('age_group', axis="columns")
y = train['age_group']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
model = RandomForestClassifier(n_estimators=20)
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
test.head()
```

```python
imputer = SimpleImputer(strategy='median')
test[test.columns] = imputer.fit_transform(test)

test.isnull().sum()
```

```python
y_pred = model.predict(test)

y_pred
```

```python
submission = pd.DataFrame({"age_group": y_pred})
submission.to_csv("submission1.csv", index=False)
```
